```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Step 1: Load the data
file_path = 'dataset/BandraKurlaComplexMumbaiIITM.csv'  # Replace with your actual file path
data = pd.read_csv(file_path)

# Step 2: Prepare the data
data['Start Date'] = pd.to_datetime(data.iloc[:, 1])  # Convert the start date column to datetime
data.set_index('Start Date', inplace=True)  # Set Start Date as the index

# Extract the parameters (3rd column onward) and convert to numeric
parameters = data.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')

# Display data info for validation
print(data.info())

# Define a function to fit and forecast using ARIMA
def forecast_arima(series, days=7):
    # Ensure no NaN values in series
    series = series.dropna()
    
    # Skip series with insufficient data
    if len(series) < 10:  # Minimum number of points to fit ARIMA
        print(f"Skipping series with insufficient data (length = {len(series)}).")
        return np.nan * np.ones(days)
    
    try:
        # Fit ARIMA model
        model = ARIMA(series, order=(5, 1, 0))  # Adjust ARIMA parameters if necessary
        model_fit = model.fit()
        # Forecast future values
        forecast = model_fit.forecast(steps=days)
        return forecast
    except Exception as e:
        print(f"ARIMA model failed: {e}")
        return np.nan * np.ones(days)

# Predict and plot for each parameter
predictions = {}
for param in parameters.columns:
    series = parameters[param]
    forecast = forecast_arima(series, 7)

    # Store predictions
    predictions[param] = forecast

    # Plot the results if forecast is successful
    if not np.isnan(forecast).all():
        plt.figure(figsize=(10, 5))
        plt.plot(series.index, series, label='Actual')
        
        # Generate future dates
        future_dates = pd.date_range(series.index[-1] + pd.Timedelta(days=1), periods=7)
        plt.plot(future_dates, forecast, label='Predicted', linestyle='--')
        plt.xlabel('Date')
        plt.ylabel(param)
        plt.title(f'{param} Prediction')
        plt.legend()
        plt.show()

# Save predictions to CSV
pred_df = pd.DataFrame(predictions, index=future_dates)
pred_df.to_csv('predictions.csv')
print("Predictions saved to predictions.csv.")
```
